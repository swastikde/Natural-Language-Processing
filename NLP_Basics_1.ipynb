## $Tokenization$

```python
corpus = "Hello, how are you doing today?\
      I hope everything is going well for you.\
        Have a great day!"
```

```python
### Tokenlization
### Sentense --> Paragraphs
from nltk.tokenize import sent_tokenize

sentences = sent_tokenize(corpus)
print(type(sentences))
print(len(sentences))
print(sentences)
```

```python
### Paragraphs --> Words
from nltk.tokenize import word_tokenize
words = word_tokenize(corpus)
print(type(words))
print(len(words))
print(words)
```

```python
## Paragraph --> Sentences --> Words
for sentence in sentences:
    words = word_tokenize(sentence)
    print(words)
```

```python
## Word Punct Tokenizer
from nltk.tokenize import wordpunct_tokenize
words = wordpunct_tokenize(corpus)
print(type(words))
print(len(words))
print(words)
```

```python
### Tree Bank Word Tokenizer
from nltk.tokenize import TreebankWordTokenizer
tokenizer = TreebankWordTokenizer()
words = tokenizer.tokenize(corpus)
print(type(words))
print(len(words))
print(words)
```

### Different types of tokenization
- Sentence Tokenization
- Word Tokenization
- Word Punct Tokenization
- Tree Bank Word Tokenizer

## What is Stemming and Lemmatization?
### Stemming
Stemming is the process of reducing a word to its base or root form. The resulting stem may not be a valid word in the language, but it is often sufficient for certain NLP tasks. For example, the words "running", "runner", and "ran" can all be reduced to the stem "run". Stemming is typically done using algorithms that apply a set of rules to remove common suffixes from words.
### Lemmatization
Lemmatization is the process of reducing a word to its base or dictionary form, known as a lemma. Unlike stemming, lemmatization takes into account the context and the part of speech of the word to ensure that the resulting lemma is a valid word in the language. For example, the words "running", "runner", and "ran" would all be reduced to the lemma "run". Lemmatization is typically done using a dictionary-based approach, where the algorithm looks up the word in a lexicon to find its lemma based on its part of speech.

```python
### Stemming 
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
words = ["running","ran","writer","writing","written","history","historical"]
for word in words:
    print(stemmer.stem(word))
```

```python
stemmer.stem("Congratulations")
```

## RegexStemming
### What is RegexStemming?
RegexStemming is a simple stemming technique that uses regular expressions to remove common suffixes from words. It is a rule-based approach that can be effective for certain languages and applications, but it may not be as accurate as more sophisticated stemming algorithms like Porter Stemmer or Snowball Stemmer. RegexStemming can be useful for quick and simple stemming tasks, but it may not be suitable for more complex NLP applications that require a deeper understanding of the language and its morphology.

```python
## RegexStemming Example
from nltk.stem import RegexpStemmer
stemmer = RegexpStemmer('ing$|s$|e$|able$|ible$|ment$|ness$|ful$|less$|ly$|al$|')
words = ["running","ran","writer","writing","written","history","historical"]
for word in words:
    print(stemmer.stem(word))
    
```

## What is Snowball Stemmer?
Snowball Stemmer is an improved version of the Porter Stemmer that was developed by Martin Porter in 2001. It is a more aggressive stemming algorithm that is designed to handle a wider range of languages and word forms. Snowball Stemmer uses a more comprehensive set of rules to remove suffixes from words, and it is often more effective at reducing words to their base forms than the Porter Stemmer. Snowball Stemmer is available in the NLTK library and can be used for stemming in various languages, including English, French, German, and Spanish.

```python
## Snowball Stemmer
from nltk.stem import SnowballStemmer
SnowballStemm = SnowballStemmer("english")
words = ["running","ran","writer","writing","written","history","historical"]
for word in words:
    print(word+" --> "+SnowballStemm.stem(word))
```

```python
SnowballStemm.stem("Fairly")
```

## What is Lammentization?
Lemmatization is the process of reducing a word to its base or dictionary form, known as a lemma. Unlike stemming, lemmatization takes into account the context and the part of speech of the word to ensure that the resulting lemma is a valid word in the language. For example, the words "running", "runner", and "ran" would all be reduced to the lemma "run". Lemmatization is typically done using a dictionary-based approach, where the algorithm looks up the word in a lexicon to find its lemma based on its part of speech. Lemmatization is often used in NLP tasks that require a deeper understanding of the language, such as sentiment analysis, machine translation, and information retrieval.

```python
## lamentization
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
words = ["running","ran","writer","writing","written","history","historical"]
for word in words:    print(word+" --> "+lemmatizer.lemmatize(word,pos="n"))
```
