### Apply SnowbAll Stemmer

```python
from nltk.stem import SnowballStemmer
snowballstem = SnowballStemmer("english")
```

```python
for i in range(len(sentences)):
    words = nltk.word_tokenize(sentences[i])
    words = [snowballstem.stem(word) for word in words if not word in stopwords.words("english")]
    sentences[i] = " ".join(words)
sentences
```

## Apply Wordnet Lemmatizer

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
for i in range(len(sentences)):
    words = nltk.word_tokenize(sentences[i])
    words = [lemmatizer.lemmatize(word,pos='v') for word in words if not word in stopwords.words("english")]
    sentences[i] = " ".join(words)
sentences
```
