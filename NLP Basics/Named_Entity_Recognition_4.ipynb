## What is Named Entity Recognition (NER)?
Named Entity Recognition (NER) is a subtask of Natural Language Processing (NLP) thatinvolves identifying and classifying named entities in text into predefined categories such as person names, organizations, locations, dates, and more. The goal of NER is to extract meaningful information from unstructured text and convert it into structured data that can be easily analyzed and utilized in various applications.

```python
sentence="The Eiffel Tower was built from 1887 to 1889 by Gustave Eiffel, whose company specialized in building metal frameworks and structures."
```

```python
import nltk
import numpy as np
from IPython.display import display
words=nltk.word_tokenize(sentence)
tag_elements=nltk.pos_tag(words)
```

```python
chunked_tree = nltk.ne_chunk(tag_elements)
display(chunked_tree)
```
